```python
# Add the magic command "%"
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Create bar chat with pyber_ride_df: Option 1
pyber_ride_df.plot.bar(x= "Month", y= "Avg. Fare ($USD)")
plt.show()
```

```python
# Create bar chat with pyber_ride_df: Option 2
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```

```python
# Skill Drill
## Using the data from PyBer ride DataFrame, create a bar chart with the following annotations:

# 1) Add error bars.
# 2) Add caps to the error bars.
# 3) Change the color of the bars to sky blue
# 4) Rotate the labels on the x-axis to horizontal
# 5) Set the y-axis increment to every $5.

# Create x, y axes variables
x_axis = pyber_ride_df["Month"]
y_axis = pyber_ride_df["Avg. Fare ($USD)"]

# Get the stdev of pyber_ride_df y_axis
import statistics
stdev = statistics.stdev(y_axis)
stdev

# Add error bars and plot bar graph with caped yerr = stdev
plt.subplots(figsize = (8, 6))
plt.bar(x_axis, y_axis, 
        color = "skyblue",
        yerr= stdev, 
        capsize = 3, 
        label = "Avg Fare ($USD)")
plt.xlabel("Month")
plt.ylabel("Avg. Fare ($USD)")
plt.yticks(np.arange(0, 51, step=5.0))
plt.xticks(rotation=0)
plt.legend(loc ='upper right')
plt.show()
```
