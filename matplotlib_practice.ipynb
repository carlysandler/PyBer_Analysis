```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pyber Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Change figure attributes using ax approach[color = green, marker = diamond, linewidth = 2, legend for city of Boston]
fig = plt.figure()
ax = fig.add_subplot()
#ax.plot() = plt.plot() for changing fig attributes
ax.plot(x_axis, y_axis, marker = 'D', color = 'g', linewidth = 2, label = 'Boston')
# Create lables for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
#Add the legend.
ax.legend()
```

```python
# Adjust size of markers. Iterate thru y-axis, multiply each data point by 3, and add it to a new list.
y_axis_larger = []
for data in y_axis :
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Adjust size using list comprehension
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
# Skill Drill: Use object-oriented approach to make changes to bubble chart.
fig, ax = plt.subplots()
# Make bubbles 20% transparent, sky blue, 5x size data pt, add a black edge to circles
ax.scatter(y_axis, x_axis, c = "skyblue", edgecolors = 'k', alpha =0.8,
           s = [i * 5 for i in y_axis], linewidth = 2, label='Boston')
ax.invert_yaxis()
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("Pyber Fare by Month")
ax.legend()
#
                                
```
